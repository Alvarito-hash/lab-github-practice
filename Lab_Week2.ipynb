```python
def greatest(lst):
    greatest_so_far = lst[0]
    
    for item in lst:
        if item > greatest_so_far:
            greatest_so_far = item
    
    return greatest_so_far
test_greatest(greatest)
```

```python
def sum_all(lst):
    total = 0
    for item in lst:
        total += item
    return total
test_sum(sum_all)
```

```python
def mult_all(lst):
    product = 1
    for item in lst:
        product *= item
    return product
```

```python
def oper_all(arr, oper = "*"):
    if oper == "+":
        result = 0
        for item in arr:
            result += item
    else:
        result = 1
        for item in arr:
            result *= item
    return result

test_operations(oper_all)
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

test_factorial(factorial)
```

```python
def unique(lst_un):
    result = []
    for item in lst_un:
        found = False
        for r in result:
            if item == r:
                found = True
                break
        if not found:
            result.append(item)
    return result

test_unique(unique)
```

```python
def mode_counter(arr):
    elements = []   
    frequencies = []  

    for item in arr:
        found = False
        for i in range(len(elements)):
            if elements[i] == item:
                frequencies[i] += 1
                found = True
                break
        if not found:
            elements.append(item)
            frequencies.append(1)

    max_index = 0
    for i in range(1, len(frequencies)):
        if frequencies[i] > frequencies[max_index]:
            max_index = i

    return elements[max_index]

test_mode(mode_counter)
```

```python
def st_dev(list_sd):
    total = 0
    count = 0
    for item in list_sd:
        total += item
        count += 1
    mean = total / count

    variance_sum = 0
    for item in list_sd:
        diff = item - mean
        variance_sum += diff * diff
    variance = variance_sum / count

    guess = variance / 2 if variance > 1 else 1  
    for _ in range(10):  
        guess = (guess + variance / guess) / 2

    return guess

test_stdev(st_dev)
```

```python
def pangram(string):
    string = string.lower()
    
    for char_code in range(ord('a'), ord('z') + 1):
        letter = chr(char_code)
        
        found = False
        for ch in string:
            if ch == letter:
                found = True
                break
        
        if not found:
            return False
    
    return True

test_pangram(pangram)
```

```python
def sort_alpha(string):
    words = []
    current_word = ""
    for char in string:
        if char == ",":
            words.append(current_word)
            current_word = ""
        else:
            current_word += char
    words.append(current_word)  

    words_sorted = sorted(words)

    result = ""
    for i in range(len(words_sorted)):
        result += words_sorted[i]
        if i != len(words_sorted) - 1:
            result += ","
    
    return result

test_alpha(sort_alpha)
```

```python
def check_pass(password):
    if len(password) < 8:
        return False

    has_lower = False
    has_upper = False
    has_number = False
    has_special = False

    for char in password:
        if 'a' <= char <= 'z':
            has_lower = True
        elif 'A' <= char <= 'Z':
            has_upper = True
        elif '0' <= char <= '9':
            has_number = True
        else:
            has_special = True

    return has_lower and has_upper and has_number and has_special

test_pass(check_pass)
```
