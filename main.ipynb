```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import math

# Bernoulli's Distribution
p = 60 / 100  
q = 40 / 100   

print("p =", p)
print("q =", q)

# What is the probability that the first 5 fruits are all apples?
prob_first_5_apples = p ** 5
print("Probabilidad de que las primeras 5 sean manzanas:", prob_first_5_apples)

# What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?
prob_5_apples_15_oranges = (p ** 5) * (q ** 15)
print("Probabilidad de 5 manzanas y luego 15 naranjas:", prob_5_apples_15_oranges)

#Binomial Distribution
# Parameters
n = 20
p = 0.6   # probability of apple
q = 1 - p # probability of orange

# Probability that the sample contains exactly 5 apples (order doesn't matter)
# Formula: P(X = 5) = C(n,5) * p^5 * q^(n-5)
prob_exact_5 = math.comb(n, 5) * (p**5) * (q**(n-5))
print("P(X = 5) =", prob_exact_5)

# Probability that less than 5 apples are picked (CDF from k=0 to 4)
# Formula: P(X < 5) = sum C(n,k) * p^k * q^(n-k)
prob_less_than_5 = sum(math.comb(n, k) * (p**k) * (q**(n-k)) for k in range(5))
print("P(X < 5) =", prob_less_than_5)

# PMF plot for the binomial distribution
x_values = list(range(n + 1))
pmf_values = [math.comb(n, k) * (p**k) * (q**(n-k)) for k in x_values]

plt.figure(figsize=(10, 5))
markerline, stemlines, baseline = plt.stem(x_values, pmf_values)
plt.setp(stemlines, linewidth=1)
plt.setp(markerline, markersize=6)

plt.xlabel("Number of apples (k)")
plt.ylabel("P(X = k)")
plt.title("PMF of Binomial Distribution (n=20, p=0.6)")
plt.tight_layout()
plt.show()

# What is the probability that the number of goals scored in a match is 5?

lam = 2.3   # media de goles ()
k = 5       # nmero de goles que queremos calcular

# Frmula de Poisson: P(X=k) = (e^- * ^k) / k!
prob_5 = (math.exp(-lam) * lam**k) / math.factorial(k)

print(f"Probabilidad de marcar exactamente 5 goles: {prob_5}")

# Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.

k_values = list(range(11))  # goles de 0 a 10

poisson_probs = [
    (math.exp(-lam) * lam**k) / math.factorial(k)
    for k in k_values
]

# Grfico
plt.figure(figsize=(8, 5))
plt.plot(k_values, poisson_probs, marker='o')
plt.xlabel("Nmero de goles (k)")
plt.ylabel("Probabilidad P(k)")
plt.title("Distribucin de Poisson para goles ( = 2.3)")
plt.grid(True)

plt.show()
```
