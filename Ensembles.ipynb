```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
X = df.drop(columns="target")
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)

y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)

print(f"Decision Tree Train Accuracy: {accuracy_score(y_train, y_train_pred):.4f}")
print(f"Decision Tree Test Accuracy: {accuracy_score(y_test, y_test_pred):.4f}")
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

base_tree = DecisionTreeClassifier(max_depth=1, random_state=0)

bagging = BaggingClassifier(estimator=base_tree, n_estimators=100, random_state=0)
bagging.fit(X_train, y_train)

y_train_pred_bag = bagging.predict(X_train)
y_test_pred_bag = bagging.predict(X_test)

print(f"Bagging Train Accuracy: {accuracy_score(y_train, y_train_pred_bag):.4f}")
print(f"Bagging Test Accuracy: {accuracy_score(y_test, y_test_pred_bag):.4f}")
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

base_tree = DecisionTreeClassifier(max_depth=1, random_state=0)

adaboost = AdaBoostClassifier(estimator=base_tree, n_estimators=100, random_state=0)
adaboost.fit(X_train, y_train)

y_train_pred_ada = adaboost.predict(X_train)
y_test_pred_ada = adaboost.predict(X_test)

print(f"AdaBoost Train Accuracy: {accuracy_score(y_train, y_train_pred_ada):.4f}")
print(f"AdaBoost Test Accuracy: {accuracy_score(y_test, y_test_pred_ada):.4f}")
```
